# 1. 単一記事の取得
- 単一記事のURL：https://diamond-rm.net/store/105874/

```python
from newspaper import Article
```

```python
url = "https://diamond-rm.net/store/105874/"
```

```python
article = Article(url, memorize_articles=False)
```

```python
#記事の取得
article.download()
```

```python
#記事の解析
article.parse()
```

```python
#タイトルの表示
article.title
```

```python
#出版日の表示
article.publish_date
```

```python
#筆者の表示
article.authors
```

```python
#本文の表示
article.text.replace("\n", " ")
```

```python
#自然言語処理
import nltk
```

```python
#downloadには"punkt"を指定
nltk.download("punkt")
```

```python
# ->> 日本語の要約をする場合は、tinysegmenter(0.3)を適用する必要がある
!pip install tinysegmenter==0.3
```

```python
article.nlp()
```

```python
#本文の要約
article.summary
```

# 複数記事からの記事取得

```python
import newspaper
```

```python
url = "https://www.itmedia.co.jp/"
```

```python
toppage = newspaper.build(url)
```

```python
toppage.articles
```

```python
for article in toppage.articles:
  article.download()
  article.parse()
  print(article.title)
  print(article.text)
```
